```python
import numpy as np
from datetime import datetime
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib import cm


%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
DATA_FILE = 'Uber-Jan-Feb-FOIL.csv'
uber_data = pd.read_csv(DATA_FILE)
uber_data.head()
```

```python
uber_data.info()
```

```python
uber_data['date'] = pd.to_datetime(uber_data['date'], format="%m/%d/%Y")
uber_data['DayOfWeekNum'] = uber_data['date'].dt.dayofweek
uber_data['DayOfWeek'] = uber_data['date'].dt.weekday_name
uber_data['MonthDayNum'] = uber_data['date'].dt.day
```

```python
uber_weekdays = uber_data.pivot_table(index=['DayOfWeekNum','DayOfWeek'],values='dispatching_base_number',aggfunc='count')

uber_weekdays.plot(kind='bar', figsize=(8,6))
plt.ylabel('Total Journeys')
plt.title('Journeys by Week Day')
```

```python
uber_weekdays = uber_data.pivot_table(index=['DayOfWeekNum','DayOfWeek'],values='trips',aggfunc='count')

uber_weekdays.plot(kind='bar', figsize=(8,6))
plt.ylabel('Total Trips')
plt.title('Journeys by Week Day')
```

```python
uber_weekdays = uber_data.pivot_table(index=['DayOfWeekNum','DayOfWeek'],values='active_vehicles',aggfunc='count')

uber_weekdays.plot(kind='bar', figsize=(8,6))
plt.ylabel('Total Trips')
plt.title('Journeys by Week Day')
```
